```python
last_lightcurve_data = ""
last_df = None
widget_out2 = None  # to be defined later by UI
def plot_lightcurve(lightcurve_data, xmin=None, xmax=None, moving_average_window = None, subject_id = None):
    global last_lightcurve_data
    global last_df
    global widget_out2
    # process input arguments
    df = None
    if lightcurve_data.startswith('{"y"'):
        # process data into a dataframe, and cache it in last_df to make UI more responsive
        if (lightcurve_data != last_lightcurve_data):
            df = pd.read_json(lightcurve_data)
            last_lightcurve_data = lightcurve_data
            last_df = df
        else:
            df = last_df    
        # zoom in to a subset of data
        if xmin != None:
            df = df[df.x >= xmin]
        if xmax != None:
            df = df[df.x <= xmax]
    else: # case non lightcurve json data, ignore them
        df = pd.DataFrame()
    
    subject_info = f"Subject {subject_id}" if subject_id != None else ""
        
    # default plot text properties
    matplotlib.rcParams.update({'font.size':24}) 
    matplotlib.rcParams.update({'font.family':'sans-serif'})

    plt.figure(figsize=(30,12)) 

    # Misc. labels, setup etc.
    plt.gca().set_facecolor("darkslategrey")
    plt.xlabel("Days")
    plt.ylabel("Brighness", labelpad=16)    
    
    if len(df) > 0:
        y_series = df['y']
        y_label = "Brightness"
        dot_color = 'whitesmoke'
        if (moving_average_window != None):
            df['x_ts'] = df['x'].apply(lambda x: pd.Timestamp(x * 1000000000 * 60 * 60 * 24))
            y_series = df.rolling(moving_average_window, on='x_ts')['y'].mean()
            y_label = f"Brightness,  {moving_average_window} moving average"
            dot_color = 'lightyellow'

        plt.scatter(df['x'], y_series, color=dot_color, alpha=1.0, label="observation")
        plt.ylabel(y_label, labelpad=16)
        
        # draw -3 standard deviation line (of the data shown only if the user zooms in)
        xmax_shown = np.max(df['x'])    
        xmin_shown = np.min(df['x'])    
        plt.hlines(np.nanmean(df['y'])- 3 * np.nanstd(df['y']), xmin, xmax_shown, linewidth=4, color="red", label=r"$-3\sigma$")
        plt.legend()
        plt.title(f"Lightcurve - {subject_info} ({len(df)} observations)", fontdict={'fontsize': 36}, pad=24)
        
        # add a textual summary
        widget_out2.clear_output()
        with widget_out2:
            print('Brightness summary:')
            print(df['y'].describe())

    else:
        plt.title(f"Lightcurve - No data", fontdict={'fontsize': 36}, pad=24)
        widget_out2.clear_output()
        
    plt.show()
    return ""
```
